```python
import pandas as pd
from pandas import read_csv
import matplotlib.pyplot as plt
```

```python
df = read_csv("amazon.csv")
```

```python
df.head()
```

```python
# Cleaning the data
df = df.drop(["img_link", "product_link"], axis=1)
df.head()
```

```python
# getting the priduct with the highest rating
data_sorted = df.sort_values(by="rating", ascending=False)
```

```python
top_five_products = data_sorted.head(5)
```

```python
product_id = top_five_products["product_id"]
product_ratings = top_five_products["rating"]
print(product_id)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(product_id, product_ratings, color="skyblue")
plt.xlabel("Product ID")
plt.ylabel("Rating")
plt.title("Top 5 Products by Rating")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
